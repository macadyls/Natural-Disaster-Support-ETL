```python
# Dependencies
import pandas as pd
```

```python
# File paths
data_path_1 = "Resources/volcano_data_2010.csv"
data_path_2 = "Resources/Worldwide-Earthquake-database.csv"
data_path_3 = "Resources/tsunami_dataset.csv"
```

```python
# Read CSV
volcano_data = pd.read_csv(data_path_1)
worldwide_earthquake_database = pd.read_csv(data_path_2)
tsunami_dataset = pd.read_csv(data_path_3)
```

```python
volcano_data.head()
```

```python
#Creating DataFrame
cleaned_volcano_df=pd.DataFrame(volcano_data, columns=["Year",
                                                       "Month",
                                                                        "Day",
                                                                        "Country",
                                                                        "Location",                                                                        
                                                                        "Latitude",
                                                                        "Longitude",
                                                                        "Elevation", 
                                                                        "Type", 
                                                                        "Status",                                                                        
                                                                        "TOTAL_DEATHS",
                                                                        "TOTAL_MISSING",
                                                                        "TOTAL_INJURIES",
                                                                        "TOTAL_HOUSES_DESTROYED",
                                                                        "TOTAL_HOUSES_DAMAGED",
                                                                        "TOTAL_DAMAGE_MILLIONS_DOLLARS"])

cleaned_volcano_df= cleaned_volcano_df.apply(lambda x: x.str.upper() if x.dtype == "object" else x)  
cleaned_volcano_df.head()
```

```python
# Reset index and save to new CSV file
cleaned_volcano_df.reset_index(drop = True, inplace = True)
cleaned_volcano_df= cleaned_volcano_df.rename(columns={"Year":"YEAR", "Month":"MONTH", "Day":"DAY", "Country":"COUNTRY", "location":"LOCATION", "Latitude":"LATITUDE", "longitude":"LONGITUDE", "Elevation":"ELEVATION", "Type":"TYPE", "Status":"STATUS","TOTAL_DEATHS":"TOTAL DEATHS", "TOTAL_MISSING":"TOTAL MISSING", "TOTAL_INJURIES":"TOTAL INJURIES","TOTAL_HOUSES_DESTROYED":"TOTAL HOUSES DESTROYED", "TOTAL_HOUSES_DAMAGED":"TOTAL HOUSES DAMAGED", "TOTAL_DAMAGE_MILLIONS_DOLLARS":"TOTAL DAMAGES(MILLION $)"})
cleaned_volcano_df.to_csv("Resources/cleaned_volcano_data.CSV")
cleaned_volcano_df
```

```python
worldwide_earthquake_database.head()
```

```python
#Creating DataFrame
cleaned_worldwide_earthquake_df=pd.DataFrame(worldwide_earthquake_database, columns=["YEAR",
                                                                        "MONTH",
                                                                        "DAY",
                                                                        "HOUR",
                                                                        "MINUTE",
                                                                        "SECOND",
                                                                        "FLAG_TSUNAMI",
                                                                        "FOCAL_DEPTH",
                                                                        "EQ_PRIMARY",                                                                    
                                                                        "INTENSITY",
                                                                        "COUNTRY",
                                                                        "LOCATION_NAME",                                                                        
                                                                        "LATITUDE",
                                                                        "LONGITUDE",                                                                        
                                                                        "TOTAL_DEATHS",
                                                                        "TOTAL_MISSING",
                                                                        "TOTAL_INJURIES",
                                                                        "TOTAL_HOUSES_DESTROYED",
                                                                        "TOTAL_HOUSES_DAMAGED",
                                                                        "TOTAL_DAMAGE_MILLIONS_DOLLARS"])
cleaned_worldwide_earthquake_df.head()
```

```python
# Filter the dataset for years between 2010-2018 and reset index and save to new CSV file
cleaned_worldwide_earthquake_df = cleaned_worldwide_earthquake_df.loc[(cleaned_worldwide_earthquake_df)["YEAR"] > 2009]
cleaned_worldwide_earthquake_df = cleaned_worldwide_earthquake_df.loc[(cleaned_worldwide_earthquake_df)["YEAR"] < 2019]
cleaned_worldwide_earthquake_df.reset_index(drop = True, inplace = True)
cleaned_worldwide_earthquake_df= cleaned_worldwide_earthquake_df.rename(columns={"FLAG_TSUNAMI": "TSUNAMI", "FOCAL_DEPTH": "FOCAL DEPTH", "LOCATION_NAME":"LOCATION", "TOTAL_DEATHS":"TOTAL DEATHS", "TOTAL_MISSING":"TOTAL MISSING", "TOTAL_INJURIES":"TOTAL INJURIES","TOTAL_HOUSES_DESTROYED":"TOTAL HOUSES DESTROYED", "TOTAL_HOUSES_DAMAGED":"TOTAL HOUSES DAMAGED", "TOTAL_DAMAGE_MILLIONS_DOLLARS":"TOTAL DAMAGES(MILLION $)"})
#cleaned_worldwide_earthquake_df=cleaned_worldwide_earthquake_df.replace("NaN", 0)

cleaned_worldwide_earthquake_df.to_csv("Resources/cleaned_worldwide_earthquake_data.CSV")
cleaned_worldwide_earthquake_df
```

```python
tsunami_dataset.head()
```

```python
#Creating DataFrame
cleaned_tsunami_df=pd.DataFrame(tsunami_dataset, columns=["YEAR",
                                                                        "MONTH",
                                                                        "DAY",
                                                                        "HOUR",
                                                                        "MINUTE",
                                                                        "COUNTRY",
                                                                        "LOCATION_NAME",                                                                        
                                                                        "LATITUDE",
                                                                        "LONGITUDE",
                                                                        "CAUSE"])
cleaned_tsunami_df.head()
```

```python
# Filter the dataset for years between 2010-2018 and reset index and save to new CSV file
cleaned_tsunami_df = cleaned_tsunami_df.loc[(cleaned_tsunami_df)["YEAR"] > 2009]
cleaned_tsunami_df = cleaned_tsunami_df.loc[(cleaned_tsunami_df)["YEAR"] < 2019]
cleaned_worldwide_earthquake_df= cleaned_worldwide_earthquake_df.rename(columns={"LOCATION_NAME":"LOCATION"})
cleaned_tsunami_df.reset_index(drop = True, inplace = True)
cleaned_tsunami_df.to_csv("Resources/cleaned_tsunami_data.CSV")
cleaned_tsunami_df
```
